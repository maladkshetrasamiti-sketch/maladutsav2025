```python
def status_badge(status):
    if not status or status.strip() == "":
        status = "Not Called"

    s = status.lower()

    if "whatsapp" in s:
        return '<span class="badge badge-wa">WhatsApp</span>'
    if "confirmed" in s:
        return '<span class="badge badge-confirmed">Confirmed</span>'
    if "maybe" in s:
        return '<span class="badge badge-maybe">Maybe</span>'
    if "busy" in s:
        return '<span class="badge badge-busy">Phone Busy</span>'
    if "not reachable" in s:
        return '<span class="badge badge-nr">Not Reachable</span>'
    if "not coming" in s:
        return '<span class="badge badge-no">Not Coming</span>'
    if "not called" in s:
        return '<span class="badge badge-notcalled">Not Called</span>'

    # fallback
    return f'<span class="badge badge-notcalled">{status}</span>'
```

```python
def row_color_for_status(status):
    s = (str(status or "")).strip().lower()
    # subtle pastel tints
    if "confirmed" in s:
        return "#e6fff4"   # light green
    if "whatsapp" in s:
        return "#f0fff9"   # palest green
    if "maybe" in s:
        return "#fff9e6"   # light amber
    if "busy" in s or "phone busy" in s:
        return "#fff6ec"   # light orange
    if "not reachable" in s:
        return "#fff2f2"   # light red/pink
    if "not coming" in s or "not coming" == s:
        return "#f3f4f6"   # light grey
    if "not called" in s or s == "":
        return "#fff7f7"   # pale rosy for pending
    # default: no tint
    return "transparent"
```
