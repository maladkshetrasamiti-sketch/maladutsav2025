```python
def status_badge(status):
    if not status or status.strip() == "":
        status = "Not Called"

    s = status.lower()

    if "whatsapp" in s:
        return '<span class="badge badge-wa">WhatsApp</span>'
    if "confirmed" in s:
        return '<span class="badge badge-confirmed">Confirmed</span>'
    if "maybe" in s:
        return '<span class="badge badge-maybe">Maybe</span>'
    if "busy" in s:
        return '<span class="badge badge-busy">Phone Busy</span>'
    if "not reachable" in s:
        return '<span class="badge badge-nr">Not Reachable</span>'
    if "not coming" in s:
        return '<span class="badge badge-no">Not Coming</span>'
    if "not called" in s:
        return '<span class="badge badge-notcalled">Not Called</span>'

    # fallback
    return f'<span class="badge badge-notcalled">{status}</span>'
```

```python
#!/usr/bin/env python3
# refactored_generator.py
from fileinput import filename
import pandas as pd
from datetime import datetime
import os
import urllib.parse
import pdfkit
import re

# === CONFIG ===
input_csv = karykarta_assignment_list   # CSV file with data
output_folder = karykarta_assignment_list_output_folder    # Folder for HTML output

# === CREATE OUTPUT FOLDER IF NOT EXISTS ===
os.makedirs(output_folder, exist_ok=True)

# --- Write shared CSS file (styles.css) into output_folder ---
css_content = """
/* ====== styles.css (generated) ====== */
body { font-family: Arial, sans-serif; margin: 20px; background: #f8f9fa; }
h2 { text-align: center; }
.top-note { font-family: Arial, sans-serif; color: #333; line-height: 1.4; padding: 8px; max-width: 100%; box-sizing: border-box; font-size: 14px; background: #fff; border-radius: 8px; }
.table-wrapper { width: 100%; overflow-x: auto; -webkit-overflow-scrolling: touch; background: #fff; padding: 8px; border-radius: 8px; box-shadow: 0 0 6px rgba(0,0,0,0.06); }
table { width: 100%; border-collapse: collapse; margin-top: 12px; table-layout: auto; white-space: nowrap; }
th, td { border: 1px solid #ccc; padding: 8px; text-align: left; vertical-align: top; }
th { background-color: #f2f2f2; position: sticky; top: 0; z-index: 1; }
a { color: #007bff; text-decoration: none; }
a:hover { text-decoration: underline; }
#searchInput {
    width: 100%;
    max-width: 560px;
    font-size: 1em;
    padding: 10px 14px;
    border-radius: 25px;
    border: 2px solid #007bff;
    margin: 12px auto;
    display: block;
    box-sizing: border-box;
    background: white;
}
mark { background: yellow; color: black; font-weight: bold; padding: 0; }
.no-results { display: none; text-align: center; color: #777; margin-top: 12px; }

@media (max-width: 600px) {
    th, td { padding: 6px 8px; font-size: 0.95em; }
}

/* Badges */
.badge {
    padding: 3px 10px;
    border-radius: 12px;
    color: white;
    font-size: 12px;
    display: inline-block;
    font-weight: 600;
}
.badge-wa { background: #25D366; }          /* WhatsApp Green */
.badge-confirmed { background: #007bff; }   /* Blue */
.badge-maybe { background: #f0ad4e; }       /* Amber */
.badge-busy { background: #ff9800; }        /* Orange */
.badge-nr { background: #dc3545; }          /* Red */
.badge-no { background: #6c757d; }          /* Grey */
.badge-notcalled { background: #b30000; }   /* Dark Red */

/* Floating Action Bar (bottom-right) */
.fab-bar {
  position: fixed;
  right: 14px;
  bottom: 14px;
  display: flex;
  flex-direction: column;
  gap: 8px;
  z-index: 9999;
  align-items: center;
}

/* Individual circular buttons */
.fab {
  width: 48px;
  height: 48px;
  border-radius: 50%;
  background: #ffffff;
  box-shadow: 0 6px 14px rgba(0,0,0,0.12);
  display: inline-flex;
  align-items: center;
  justify-content: center;
  cursor: pointer;
  border: 0;
  font-size: 18px;
  transition: transform .12s ease, box-shadow .12s ease;
  padding: 0;
}
.fab:active { transform: scale(.96); }
.fab:hover { box-shadow: 0 10px 20px rgba(0,0,0,0.14); }

/* colored buttons for filters */
.fab-filter { background: #f8f9fa; color: #333; }
.fab-filter.active { box-shadow: 0 8px 18px rgba(0,0,0,0.18); transform: translateY(-2px); }

/* larger primary button for 'Go to Top' */
.fab-top { background: #007bff; color: white; width:56px; height:56px; font-weight:700; }

/* Status filter toolbar */
.status-toolbar {
  display:flex;
  gap:8px;
  flex-wrap:wrap;
  align-items:center;
  margin: 10px 0;
  padding: 6px;
}
.status-filter-btn {
  display:inline-flex;
  align-items:center;
  gap:8px;
  padding:6px 10px;
  border-radius:20px;
  background:#fff;
  box-shadow:0 4px 10px rgba(0,0,0,0.06);
  cursor:pointer;
  border:1px solid rgba(0,0,0,0.04);
  font-weight:600;
  font-size:13px;
}
.status-filter-btn .count {
  min-width:26px;
  text-align:center;
  padding:2px 6px;
  border-radius:12px;
  font-size:12px;
  color:#222;
  background:rgba(0,0,0,0.05);
}
.status-filter-btn.active {
  box-shadow:0 8px 18px rgba(0,0,0,0.12);
  transform: translateY(-2px);
  border-color: rgba(0,0,0,0.12);
}
.status-toolbar .legend { font-size:13px; color:#444; margin-left:8px; }
@media (max-width:640px){
  .status-filter-btn { padding:8px 10px; font-size:13px; }
}
"""

css_path = os.path.join(output_folder, "styles.css")
with open(css_path, "w", encoding="utf-8") as f:
    f.write(css_content)

# --- Write shared JS file (scripts.js) into output_folder ---
# Use raw string to avoid Python escape warnings
js_content = r"""
/* ====== scripts.js (generated) ====== */

/* Helper: escape regex special chars */
function escapeRegex(s) {
    return s.replace(/[.*+?^${}()|[\]\\]/g, '\\$&');
}

/* Save original innerHTML of each td once (so we can restore) */
(function cacheCellOriginalsOnce() {
    document.addEventListener('DOMContentLoaded', function() {
        const tds = document.querySelectorAll('#dataTable tbody td');
        tds.forEach(td => {
            if (!td.dataset.original) td.dataset.original = td.innerHTML;
        });
    });
})();

/* Highlight matches in text nodes under an element, preserving tags (links etc) */
function highlightUnderElement(el, term) {
    if (!term) return;
    const regex = new RegExp(escapeRegex(term), 'gi');

    function walk(node) {
        if (node.nodeType === Node.TEXT_NODE) {
            const text = node.nodeValue;
            const match = regex.exec(text);
            if (match) {
                regex.lastIndex = 0; // reset for subsequent calls
                const frag = document.createDocumentFragment();
                let lastIndex = 0;
                let m;
                while ((m = regex.exec(text)) !== null) {
                    const before = text.slice(lastIndex, m.index);
                    if (before) frag.appendChild(document.createTextNode(before));
                    const mark = document.createElement('mark');
                    mark.textContent = m[0];
                    frag.appendChild(mark);
                    lastIndex = m.index + m[0].length;
                }
                const after = text.slice(lastIndex);
                if (after) frag.appendChild(document.createTextNode(after));
                node.parentNode.replaceChild(frag, node);
            }
        } else if (node.nodeType === Node.ELEMENT_NODE && node.tagName.toLowerCase() !== 'script' && node.tagName.toLowerCase() !== 'style') {
            const children = Array.from(node.childNodes);
            for (const child of children) walk(child);
        }
    }

    walk(el);
}

/* Global active status filter (null = no status filter) */
let activeStatusFilter = null;

/* Helper: normalize status text for comparison */
function normStatusText(s) {
  if (!s) return '';
  return s.toString().trim().toLowerCase();
}

/* Toggle status filter: same button toggles on/off */
function toggleStatusFilter(statusKey) {
  if (activeStatusFilter === statusKey) activeStatusFilter = null;
  else activeStatusFilter = statusKey;

  // visually update FABs if they exist
  const notBtn = document.getElementById('filterNotCalledBtn');
  const whatsBtn = document.getElementById('filterWhatsAppBtn');
  if (notBtn) notBtn.classList.toggle('active', activeStatusFilter === 'not called');
  if (whatsBtn) whatsBtn.classList.toggle('active', activeStatusFilter === 'whatsapp');

  searchTable();
}

/* Clear highlights + clear active filters + clear search input */
function clearHighlightsAndSearch() {
  activeStatusFilter = null;
  const notBtn = document.getElementById('filterNotCalledBtn');
  const whatsBtn = document.getElementById('filterWhatsAppBtn');
  if (notBtn) notBtn.classList.remove('active');
  if (whatsBtn) whatsBtn.classList.remove('active');

  const input = document.getElementById('searchInput');
  if (input) input.value = '';

  const tds = document.querySelectorAll('#dataTable tbody td');
  tds.forEach(td => {
    if (td.dataset && td.dataset.original !== undefined) {
      td.innerHTML = td.dataset.original;
    }
  });

  const rows = document.querySelectorAll('#dataTable tbody tr');
  rows.forEach(r => r.style.display = '');
  const nr = document.getElementById('noResults');
  if (nr) nr.style.display = 'none';
}

/* Simple go-to-top */
function goToTop() {
  window.scrollTo({ top: 0, behavior: 'smooth' });
}

/* searchTable() that respects activeStatusFilter and highlights */
function searchTable() {
  const input = document.getElementById('searchInput');
  const filter = (input && input.value || '').trim().toLowerCase();
  const rows = document.querySelectorAll('#dataTable tbody tr');
  let foundAny = false;

  rows.forEach(row => {
    const tds = row.querySelectorAll('td');
    // restore original
    tds.forEach(td => {
      if (td.dataset && td.dataset.original !== undefined) {
        td.innerHTML = td.dataset.original;
      }
    });

    const rowText = (row.textContent || '').toLowerCase();
    const textMatch = !filter || rowText.includes(filter);

    // status match
    let statusMatch = true;
    if (activeStatusFilter) {
      statusMatch = false;
      const statusCandidates = Array.from(row.querySelectorAll('td')).map(td => (td.textContent || '').toLowerCase());
      for (const cellText of statusCandidates) {
        if (activeStatusFilter === 'not called') {
          if (cellText.includes('not called')) { statusMatch = true; break; }
        } else if (activeStatusFilter === 'whatsapp') {
          if (cellText.includes('whatsapp')) { statusMatch = true; break; }
        } else {
          if (cellText.includes(activeStatusFilter)) { statusMatch = true; break; }
        }
      }
    }

    const isMatch = textMatch && statusMatch;
    row.style.display = isMatch ? '' : 'none';

    if (isMatch && filter) {
      tds.forEach(td => {
        if ((td.textContent || '').toLowerCase().includes(filter)) {
          highlightUnderElement(td, filter);
        }
      });
    }

    if (isMatch) foundAny = true;
  });

  const nr = document.getElementById('noResults');
  if (nr) nr.style.display = (foundAny || !filter ? 'none' : 'block');

  // update counts after filtering
  if (typeof updateStatusCounts === 'function') {
    setTimeout(updateStatusCounts, 40);
  }
}

/* Helper: read normalized status text from a row */
function getStatusFromRow(row) {
  const badge = row.querySelector('span.badge');
  if (badge && badge.textContent) return normStatusText(badge.textContent);
  const text = normStatusText(row.textContent || '');
  const keys = ['not called','whatsapp','confirmed','maybe','phone busy','busy','not reachable','not coming'];
  for (const k of keys) if (text.includes(k)) return k;
  return '';
}

/* Update counts for toolbar */
function updateStatusCounts() {
  const rows = Array.from(document.querySelectorAll('#dataTable tbody tr'));
  const counts = {'not called':0,'whatsapp':0,'confirmed':0,'maybe':0,'busy':0,'not reachable':0,'not coming':0};
  let visible = 0;
  rows.forEach(row => {
    const s = getStatusFromRow(row);
    if (s.includes('confirmed')) counts['confirmed']++;
    else if (s.includes('maybe')) counts['maybe']++;
    else if (s.includes('whatsapp')) counts['whatsapp']++;
    else if (s.includes('not called')) counts['not called']++;
    else if (s.includes('busy') || s.includes('phone busy')) counts['busy']++;
    else if (s.includes('not reachable')) counts['not reachable']++;
    else if (s.includes('not coming')) counts['not coming']++;

    const cs = window.getComputedStyle(row);
    if (cs.display !== 'none' && cs.visibility !== 'hidden' && row.offsetParent !== null) visible++;
  });

  // write counts to DOM safely
  const safeSet = (id, val) => { const el = document.getElementById(id); if (el) el.textContent = val; };
  safeSet('count-notcalled', counts['not called']);
  safeSet('count-whatsapp', counts['whatsapp']);
  safeSet('count-confirmed', counts['confirmed']);
  safeSet('count-maybe', counts['maybe']);
  safeSet('count-busy', counts['busy']);
  safeSet('count-nr', counts['not reachable']);
  safeSet('count-no', counts['not coming']);

  safeSet('totalVisible', visible);
  safeSet('totalAll', rows.length);
}

/* toolbar integration helper */
function toggleStatusFilterFromToolbar(statusKey) {
  if (activeStatusFilter === statusKey) activeStatusFilter = null;
  else activeStatusFilter = statusKey;

  document.querySelectorAll('.status-filter-btn').forEach(btn => {
    const k = btn.getAttribute('data-status-key');
    btn.classList.toggle('active', activeStatusFilter === k);
  });

  if (typeof searchTable === 'function') searchTable();
  else applyStatusFilterDirectly();
  setTimeout(updateStatusCounts, 50);
}

function applyStatusFilterDirectly() {
  const rows = document.querySelectorAll('#dataTable tbody tr');
  rows.forEach(row => {
    if (!activeStatusFilter) { row.style.display = ''; return; }
    const s = getStatusFromRow(row);
    if ((s || '').includes(activeStatusFilter)) row.style.display = '';
    else row.style.display = 'none';
  });
  updateStatusCounts();
}

/* ensure counts on load + wire search input to counts */
document.addEventListener('DOMContentLoaded', function() {
  document.querySelectorAll('#dataTable tbody td').forEach(td => {
    if (td.dataset.original === undefined) td.dataset.original = td.innerHTML;
  });

  updateStatusCounts();

  const input = document.getElementById('searchInput');
  if (input) {
    let to; input.addEventListener('input', function() {
      clearTimeout(to);
      to = setTimeout(function() {
        if (typeof searchTable === 'function') searchTable();
        updateStatusCounts();
      }, 150);
    });
  }

  const tbody = document.querySelector('#dataTable tbody');
  if (tbody) {
    tbody.addEventListener('click', function() { setTimeout(updateStatusCounts, 40); });
  }

  // MutationObserver to auto-update counts when rows change
  (function() {
    const table = document.querySelector('#dataTable');
    if (!table) return;
    const observer = new MutationObserver(function() { setTimeout(updateStatusCounts, 50); });
    const tb = table.querySelector('tbody') || table;
    observer.observe(tb, { childList: true, subtree: true, characterData: true });
  })();
});
"""

js_path = os.path.join(output_folder, "scripts.js")
with open(js_path, "w", encoding="utf-8") as f:
    f.write(js_content)

# === READ CSV ===
df = pd.read_csv(input_csv)
# Clean column names if needed
df.columns = df.columns.str.strip()

dfCallLog = pd.read_csv(call_log_data_csv)
dfCallLog.columns = dfCallLog.columns.str.strip()
dfCallLog["Timestamp"] = pd.to_datetime(dfCallLog["Timestamp"], errors="coerce")
dfCallLog = (
    dfCallLog[dfCallLog["Program Name"] == Program_Name]
    [["Timestamp", "Phone Number", "Call Status ?", "Remarks"]]
    .sort_values("Timestamp")  # oldest -> newest
    .drop_duplicates(subset="Phone Number", keep="last")
    .copy()
)

def normalize_phone(s):
    if pd.isna(s):
        return ""
    return str(s).strip().replace(" ", "")

# remove non-digit characters (keeps only 0-9)
dfCallLog["Phone_norm"] = dfCallLog["Phone Number"].astype(str).str.replace(r"\D+", "", regex=True)
df["Mobile_norm"] = df["Mobile Number"].astype(str).str.replace(r"\D+", "", regex=True)

df = df.merge(
    dfCallLog[["Timestamp", "Phone_norm", "Call Status ?", "Remarks"]],
    left_on="Mobile_norm",
    right_on="Phone_norm",
    how="left",
    suffixes=("", "_calllog")
)

# 6) Clean up helper columns (if you don't need them)
df = df.drop(columns=["Mobile_norm", "Phone_norm"])

df.head()

# === PROCESS EACH SAMITI MEMBER ===
for samiti_member_name, group in df.groupby("Samiti Member Name"):

    filepath = html_file_name_samiti_member(samiti_member_name)

    # HTML page references the external CSS and JS files we created earlier.
    html = f"""<!doctype html>
<html lang="en">
<head>
  <meta charset="utf-8">
  <title>Calling List - {samiti_member_name}</title>
  <meta name="viewport" content="width=device-width, initial-scale=1">
  <!-- Link to external stylesheet (relative to HTML file) -->
  <link rel="stylesheet" href="{os.path.basename(css_path)}">
</head>
<body>
  <img src="{image_path}" alt="Malad Kshetra Samiti Logo" style="display:block; margin-left:auto; margin-right:auto; width:100px; height:auto;">
  <h2>Calling List - {samiti_member_name}</h2>
  <div class="top-note">
    <p style="font-weight: bold; margin-bottom: 6px; font-size: 15px;">
      ‡§ï‡•É‡§™‡§Ø‡§æ ‡§®‡§ø‡§Æ‡•ç‡§®‡§≤‡§ø‡§ñ‡§ø‡§§ ‡§∏‡§∞‡§≤ ‡§ö‡§∞‡§£‡•ã‡§Ç ‡§ï‡§æ ‡§™‡§æ‡§≤‡§® ‡§ï‡§∞‡•á‡§Ç:
    </p>
    <ul style="padding-left: 18px; margin: 0; font-size: 12px;">
      <li><strong>Mobile Number</strong> ‡§™‡§∞ ‡§ï‡•ç‡§≤‡§ø‡§ï ‡§ï‡§∞‡•á‡§Ç ‚Äî ‡§ï‡•â‡§≤ ‡§Ö‡§™‡§®‡•á ‡§Ü‡§™ ‡§°‡§æ‡§Ø‡§≤ ‡§π‡•ã ‡§ú‡§æ‡§è‡§ó‡•Ä</li>
      <li><strong>SMS</strong> ‡§™‡§∞ ‡§ï‡•ç‡§≤‡§ø‡§ï ‡§ï‡§∞‡•á‡§Ç ‚Äî ‡§Ü‡§™‡§ï‡•á ‡§Æ‡•ã‡§¨‡§æ‡§á‡§≤ ‡§∏‡•á ‡§Ü‡§Æ‡§Ç‡§§‡•ç‡§∞‡§£ ‡§∏‡§Ç‡§¶‡•á‡§∂ ‡§ú‡§æ‡§è‡§ó‡§æ</li>
      <li><strong>WhatsApp</strong> ‡§™‡§∞ ‡§ï‡•ç‡§≤‡§ø‡§ï ‡§ï‡§∞‡•á‡§Ç ‚Äî ‡§µ‡•ç‡§π‡§æ‡§ü‡•ç‡§∏‡§è‡§™ ‡§ñ‡•Å‡§≤‡•á‡§ó‡§æ ‡§î‡§∞ ‡§∏‡§Ç‡§¶‡•á‡§∂ ‡§Ö‡§™‡§®‡•á ‡§Ü‡§™ ‡§™‡•ã‡§∏‡•ç‡§ü ‡§π‡•ã ‡§ú‡§æ‡§è‡§ó‡§æ</li>
      <li>‡§Ü‡§™ ‡§∏‡§Æ‡§æ‡§ú‡§¨‡§Ç‡§ß‡•Å ‡§ï‡•ã <strong>{Program_Name}</strong> ‡§ï‡•Ä ‡§ú‡§æ‡§®‡§ï‡§æ‡§∞‡•Ä ‡§¶‡•á‡§Ç</li>
    </ul>
  </div>

  <input id="searchInput" type="search" placeholder="üîç Search by Name, Number, Registration, etc..." oninput="searchTable()" />

  <!-- Status toolbar -->
  <div class="status-toolbar" id="statusToolbar" aria-hidden="false">
    <div class="status-filter-btn" data-status-key="not called" id="btn-notcalled" onclick="toggleStatusFilterFromToolbar('not called')">
      <span class="badge badge-notcalled">Not Called</span>
      <span class="count" id="count-notcalled">0</span>
    </div>
    <div class="status-filter-btn" data-status-key="whatsapp" id="btn-whatsapp" onclick="toggleStatusFilterFromToolbar('whatsapp')">
      <span class="badge badge-wa">WhatsApp</span>
      <span class="count" id="count-whatsapp">0</span>
    </div>
    <div class="status-filter-btn" data-status-key="confirmed" id="btn-confirmed" onclick="toggleStatusFilterFromToolbar('confirmed')">
      <span class="badge badge-confirmed">Confirmed</span>
      <span class="count" id="count-confirmed">0</span>
    </div>
    <div class="status-filter-btn" data-status-key="maybe" id="btn-maybe" onclick="toggleStatusFilterFromToolbar('maybe')">
      <span class="badge badge-maybe">Maybe</span>
      <span class="count" id="count-maybe">0</span>
    </div>
    <div class="status-filter-btn" data-status-key="busy" id="btn-busy" onclick="toggleStatusFilterFromToolbar('busy')">
      <span class="badge badge-busy">Phone Busy</span>
      <span class="count" id="count-busy">0</span>
    </div>
    <div class="status-filter-btn" data-status-key="not reachable" id="btn-nr" onclick="toggleStatusFilterFromToolbar('not reachable')">
      <span class="badge badge-nr">Not Reachable</span>
      <span class="count" id="count-nr">0</span>
    </div>
    <div class="status-filter-btn" data-status-key="not coming" id="btn-no" onclick="toggleStatusFilterFromToolbar('not coming')">
      <span class="badge badge-no">Not Coming</span>
      <span class="count" id="count-no">0</span>
    </div>

    <div class="legend" style="margin-left:auto;">
      <strong>Total:</strong> <span id="totalVisible">0</span> visible / <span id="totalAll">0</span> total
    </div>
  </div>

  <div class="table-wrapper">
    <table id="dataTable">
      <thead>
        <tr>
          <th>#</th>
          <th>Assigned Person Name</th>
          <th>WahtsApp Message</th>
          <th>SMS</th>
          <th>Call Status</th>
          <th>Remarks</th>
          <th>As of</th>
          <th>Call Log Link</th>
        </tr>
      </thead>
      <tbody>
"""

    # Sort the group DataFrame by 'Assigned Person' alphabetically
    group_sorted = group.sort_values(by="Assigned Person")
    i = 1
    for _, row in group_sorted.iterrows():
        assigned_person = row["Assigned Person"]
        call_status = row["Call Status ?"]
        if pd.isna(call_status):
            call_status = "Not Called"
        as_of = row["Timestamp"] if not pd.isna(row["Timestamp"]) else ""
        remarks = row["Remarks"] if not pd.isna(row["Remarks"]) else ""

        # row_color = "#e6ffe6" if str(call_status).strip().lower() == "confirmed presence" else "#ffe6e6"
        row_color = row_color_for_status(call_status)

        # Custom WhatsApp message
        whatsapp_message = event_message_in_person.format(assigned_person, samiti_member_name)
        encoded_whatsapp_message = urllib.parse.quote_plus(whatsapp_message)

        mobile = str(row["Mobile Number"]).strip()

        encoded_person = urllib.parse.quote(assigned_person)
        name_call_link = f'<a href="tel:+91{mobile}" style="color:#007bff;" onclick="this.style.color=\'green\'">{assigned_person}</a>'

        encoded_mobile = urllib.parse.quote(mobile)
        encoded_samiti = urllib.parse.quote(samiti_member_name)

        # Google Form Prefilled Link
        call_log_link = (
            f"{form_url_base}"
            f"&entry.752647600=Called"
            f"&entry.1573933320={encoded_samiti}"
            f"&entry.1585817478={encoded_mobile}"
            f"&entry.1475012068={encoded_person}"
            f"&entry.410440017={Program_Name}"
        )

        whatsapp_log_link = (
            f"{form_url_base}"
            f"&entry.752647600=WhatsApp"
            f"&entry.1573933320={encoded_samiti}"
            f"&entry.1585817478={encoded_mobile}"
            f"&entry.1475012068={encoded_person}"
            f"&entry.410440017={Program_Name}"
        )

        # tel, sms, whatsapp links (SVG)
        tel_link = f"<a onclick=\"this.style.color='green'\" href='tel:+91{mobile}'>+91 {mobile}</a>"
        sms_link = f'<a href="sms:+91{mobile}?body={encoded_whatsapp_message}" target="_blank" onclick="this.style.color=\'green\'">SMS</a>'

        whatsapp_link = f'''
            <a href="https://wa.me/91{mobile}?text={encoded_whatsapp_message}"
            target="_blank"
            style="color:#25D366;"
            onclick="this.style.opacity='0.6'">
                <svg xmlns="http://www.w3.org/2000/svg" 
                    width="22" height="22" viewBox="0 0 32 32" 
                    fill="#25D366" style="vertical-align:middle;">
                    <path d="M16 .5C7.6.5.8 7.3.8 15.7c0 3 1 5.9 2.7 8.3L.5 31.5l8-2.9c2.3 1.3 5 2 7.7 2 8.4 0 15.2-6.8 15.2-15.2S24.4.5 16 .5zm0 27.5c-2.4 0-4.7-.7-6.7-2l-.5-.3-4.7 1.7 1.6-4.9-.3-.5c-1.5-2.1-2.3-4.5-2.3-7.1 0-6.7 5.3-12.2 12.2-12.2 6.7 0 12.2 5.5 12.2 12.2 0 6.9-5.5 12.1-12.2 12.1z"/>
                    <path d="M23.2 19.3c-.3-.1-1.7-.8-2-1-.3-.1-.6-.1-.9.1-.3.3-1 .9-1.3 1.1-.2.1-.5.2-.8.1-.3-.1-1.4-.5-2.6-1.6-1-0.9-1.6-2-1.8-2.3-.2-.3 0-.6.2-.8.2-.2.3-.3.4-.5.1-.2.1-.3.2-.5.1-.1.1-.3 0-.5-.1-.1-.9-2.1-1.2-2.9-.3-.8-.7-.7-1-.7h-.9c-.3 0-.5.1-.7.3-.2.2-1 1-1 2.4 0 1.4 1 2.8 1.2 3 .1.2 2 3.4 4.9 4.8 2.9 1.4 2.9.9 3.4.9s1.7-.5 1.9-1c.2-.5.2-.9.1-1z"/>
                </svg>
            </a>
            '''

        html += f"""
            <tr style="background-color: {row_color};">
                <td>{i}</td>
                <td>{name_call_link}<br/><small style="color:#999">{mobile}</small></td>
                <td>{whatsapp_link}  &nbsp;|&nbsp;  <a onclick="this.style.color='green'" href="{whatsapp_log_link}" target="_blank">Log</a> </td>
                <td>{sms_link}</td>
                <td>{status_badge(call_status)}</td>
                <td>{remarks}</td>
                <td>{as_of}</td>
                <td><a onclick="this.style.color='green'" href="{call_log_link}" target="_blank">Log Call</a></td>
            </tr>
        """
        i += 1

    # close tbody/table and include noResults + floating bar + link to external script
    html += r"""
      </tbody>
    </table>
  </div>

  <div id="noResults" class="no-results">No matching results found</div>

  <!-- Floating Action Bar -->
  <div class="fab-bar" aria-hidden="false" id="floatingBar">
    <button class="fab fab-top" title="Go to Top" onclick="goToTop()">‚Üë</button>
    <button id="filterNotCalledBtn" class="fab fab-filter" title="Only show Not Called" onclick="toggleStatusFilter('not called')">üî¥</button>
    <button id="filterWhatsAppBtn" class="fab fab-filter" title="Only show WhatsApp" onclick="toggleStatusFilter('whatsapp')">üí¨</button>
    <button class="fab fab-filter" title="Clear search & highlights" onclick="clearHighlightsAndSearch()">‚úñ</button>
  </div>

  <!-- Link to external script (relative path) -->
  <script src="{}"></script>
</body>
</html>
""".format(os.path.basename(js_path))

    # Save the HTML file
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(html)

print(f"‚úÖ HTML files & assets successfully created in folder: {output_folder}")
```
